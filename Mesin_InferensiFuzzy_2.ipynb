```python
temperature = np.arange(1, 31)

hotMin = [0 for i in temperature if i <= 15]
hotCentre = [((i - 15) / 10) for i in temperature if 15 < i < 25]
hotMax = [1 for i in temperature if i >= 25]
# print(len(hotMin))
# print(len(hotCentre))
# print(len(hotMax))

hots = []
hot = hots.extend(hotMin)
hot = hots.extend(hotCentre)
hot = hots.extend(hotMax)
print(hots)

    
    
warmMin = [0 for i in temperature if (i <= 7) or (i >= 25)]
warmCentre = [((i - 7) / 8) for i in temperature if 7 < i <= 15]
warmMax = [((i - 25) / -10) for i in temperature if 15 < i < 25]
# print(len(warmMin))
# print(len(warmCentre))
# print(len(warmMax))
warms = []
warm = warms.extend(warmMin)
warm = warms.extend(warmCentre)
warm = warms.extend(warmMax)
print(warms)



coldMin = [0 for i in temperature if i >= 15]
coldCentre = [(1 - ((i - 7) / 8)) for i in temperature if 7 < i < 15]
coldMax = [1 for i in temperature if i <= 7]
# print(len(coldMin))
# print(len(coldCentre))
# print(len(coldMax))
colds = []
cold = colds.extend(coldMin)
cold = colds.extend(coldCentre)
cold = colds.extend(coldMax)
print(colds)

hots21 = hots[21]
warms21 = warms[21]
colds21 = colds[21]
# print(hots21)
# print(warms21)
# print(colds21)
```

```python
people = np.arange(1, 8001)

quietMin = [0 for i in people if 500 <= i <= 8000]
quietCentre = [((i - 500) / 400) for i in people if 100 < i < 500]
quietMax = [1 for i in people if i <= 100]
# print(len(quietMin))
# print(len(quietCentre))
# print(len(quietMax))
quiets = []
quiet = quiets.extend(quietMin)
quiet = quiets.extend(quietCentre)
quiet = quiets.extend(quietMax)
print(quiets)


crowdMin = [0 for i in people if 500 <= i <= 8000]
crowdCentre = [((i - 500) / 400) for i in people if 100 < i < 500]
crowdMax = [1 for i in people if i <= 100]
# print(len(crowdMin))
# print(len(crowdCentre))
# print(len(crowdMax))
crowds = []
crowd = crowds.extend(crowdMin)
crowd = crowds.extend(crowdCentre)
crowd = crowds.extend(crowdMax)
print(crowds)
```

```python
def zadeh(hot, warm):
    answer1 = []
    answer2 = []
    answer3 = []
    
    #ZADEH == max{min(FP1(x),FP2(y)),1-FP1(x)}
    for result_warm in warm:
        for result_hot in hot:
            step1 = np.minimum(result_hot, result_warm)
            step2 = (1 - result_hot)
            step3 = np.maximum(step1, step2)
            answer1.append(step1)
            answer2.append(step2)
            answer3.append(step3)
    
    return answer1, answer2, answer3

hotORwarm = zadeh(hots, warms)
hotORwarm
```

```python
def lukas(answer, y):
    answer1 = []
    answer2 = []
    ONE = 1
    #LUKAS == (x,y)=min{1,1-FP1(x)+FP2(y))
    for result_y in y:
        for result_answer in answer:
            step1 = (1 - result_answer + result_y)
            step2 = np.minimum(1, step1)
            answer1.append(step1)
            answer2.append(step2)
    
    return answer1, answer2
```

```python
import math
def mamdani(rule1, rule2):
    answer1 = []
    
    #MAMDANI == FP1(x) * FP2(y)
    for (result_r1, result_r2) in zip(rule1, rule2):
        step1 = np.maximum(result_r1, result_r2)
        answer1.append(step1)
    
    return answer1

rule_result = mamdani(rule1[1], rule2[1])
rule_divider = np.array([rule_result]).reshape(math.floor(240000/2), 2)

rule_divider
```

```python
def supreme(rule_divider, twentyone):
    
    answer_a = []
    answer_b = []
    duasatu = np.array([twentyone])
    rule_d = np.array([rule_divider])

    for (result_twentyone, result_rule_divider) in zip(duasatu, rule_d[0][0]):
        step1 = np.minimum(result_twentyone, result_rule_divider)
        answer_a.append(step1)

        
    for (result_twentyone, result_rule_divider) in zip(duasatu, rule_d[0][1]):
        step2 = np.minimum(result_twentyone, result_rule_divider)
        answer_b.append(step2)
        
    answer = np.array([np.concatenate([answer_a, answer_b])])
    answer_max_a = np.max(answer_a)
    answer_max_b = np.max(answer_b)
    answer_max = np.array([answer_max_a, answer_max_b])
#     print(answer)
    return answer_a, answer_b, answer_max

hot_result1, hot_result2, hot_max = supreme(rule_divider, hots21)
hot_data = pd.DataFrame.from_dict({"HOT Rule 1":hot_result1, 
                                     "HOT Rule 2": hot_result2, 
                                     "Hot MAX": hot_max}, orient='index').transpose()
hot_data
```
